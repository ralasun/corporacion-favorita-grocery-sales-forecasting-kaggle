#  vs.  / autocorrelation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
train=pd.read_csv('..\\train_data_store_oil.csv')
train=train.drop('id', axis=1)
items=pd.read_csv('..\\items.csv')
train=pd.merge(train, items, how='left', on='item_nbr')
```

```python
families=train.groupby('family').groups.keys()
```

```python
maxidxdic={}
for family in families:
    maxidxbyfam=train.iloc[train.index[train.family==family]].groupby('item_nbr').count()['unit_sales'].idxmax()
    print(family, maxidxbyfam)
    maxidxdic[family]=maxidxbyfam
```

```python
train.groupby('store_nbr').sum()['unit_sales'].idxmax()
```

```python
train_44=train.iloc[train.index[train.store_nbr==44]]
train_44.groupby('family')['item_nbr'].value_counts()
```

```python
%matplotlib inline
train_itemnbr=train.iloc[train.index[train.item_nbr==457928]]
train_44_itemnbr=train.iloc[train_itemnbr.index[train_itemnbr.store_nbr==44]]

dates_itmnbr=train_44_itemnbr.date
unit_sales_itemnbr=train_44_itemnbr.unit_sales

from datetime import datetime
dateslist=[]
for i in range(len(dates_itmnbr)):
    dateslist.append(datetime.strptime(dates_itmnbr.values[i], '%Y-%m-%d').date())

import matplotlib
dates=matplotlib.dates.date2num(dateslist)

unit_sales_itemnbr

plt.figure(figsize=(30,18))
plt.plot_date(dates, unit_sales_itemnbr.values, linestyle='solid', marker=None)
plt.savefig('../dates_vs_store44_item457928_poultry.png')
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(unit_sales_itemnbr.values, lags=100)
plt.savefig('..\\autocorrelationplot_store44_457928_poultry.png') ##seasonality?
```
