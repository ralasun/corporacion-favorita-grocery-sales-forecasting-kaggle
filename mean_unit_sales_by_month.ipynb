```python
def mean_unit_sales_by_month(train_df):

    months=['01','02','03','04','05','06','07','08','09','10','11','12']
    stores=[1, 2, 3, 4, 6, 7, 8, 9, 10, 17, 18, 20, 44, 45, 46, 47, 48, 49]
    for i in range(len(months)):
        month=months[i]
        train_month=train.iloc[train.index[train['date'].apply(lambda i:i.split('-')[1]) == month ]]
        train_month=train_month.reset_index()
        train_month=train_month.drop('index', axis=1)
        
        keys=[]
        serieslist=[]
        for i in range(len(stores)):
            storenbr=stores[i]
            train_month_str=train_month.iloc[train_month.index[train_month.store_nbr==storenbr]]
            train_month_str=train_month_str.reset_index()
            train_month_str=train_month_str.drop('index', axis=1)
           
            series=train_month_str.groupby(by=['family','item_nbr']).mean()['unit_sales']
            serieslist.append(series)
            keys.append(storenbr)

            if i > 0:
                ddf=pd.concat(serieslist, axis=1, keys=keys)
        
        ddf.to_csv('..\\mean_unit_sales_{}.csv'.format(month))
        print('{}.csv is saved'.format(month))
```
